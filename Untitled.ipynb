```python
import pandas as pd; import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

Base case

```python
n=10
data = []
for i in range(1,n+1):
    print(i)
    data.append(pd.read_csv("https://raw.githubusercontent.com/javmencia/COBWEBfiles/basecase/basecase{}.csv".format(i), sep="\t"))
```

```python
basecase = []
for i in range(n):
    print(i)
    d =data[i]
    temp = np.zeros(1500)

    for a in range(len(d)):
        if d['Death'][a]==-1:
            alive = [d['Birth'][a], 1500]
        elif d['Birth'][a]==-1:
            alive = [0, d['Death'][a]+1]
        else:
            alive = [d['Birth'][a], d['Death'][a]+1]
        incr = np.ones(alive[1]-alive[0])
        for i in range(alive[0], alive[1]):
            if i<1500:
                temp[i] += 1
    basecase.append(temp)

#  for j in range(1500):
#    for k in range(len(d)):
#        temp[j] +=(j in range(d['Birth'][k],d['Death'][k]+1))
#  timestamp.append(temp)
```

```python
print(len(basecase[0]))
x = [[i in range(len(basecase[j]))] for j in range(n)] 
for i in range(10):
    plt.plot(basecase[i][:1500], color = "lightskyblue")
plt.plot([sum(values) / len(values) for values in zip(*basecase)][:1500], color ="navy", linewidth = 2, linestyle='--', label = "Average")
plt.ylim(ymin=0)


plt.title('Base case: Agent count over time to 1500 ticks')
```

Immigrants

```python
n=14
data = []
for i in range(1,n+1):
    print(i)
    data.append(pd.read_csv("https://raw.githubusercontent.com/javmencia/COBWEBfiles/immigrants/immigrant{}.csv".format(i), sep="\t"))
```

```python
immigrants = []
for i in range(n):
    print(i)
    d =data[i]
    temp = np.zeros(1500)

    for a in range(len(d)):
        if d['Death'][a]==-1:
            alive = [d['Birth'][a], 1500]
        elif d['Birth'][a]==-1:
            alive = [0, d['Death'][a]+1]
        else:
            alive = [d['Birth'][a], d['Death'][a]+1]
        incr = np.ones(alive[1]-alive[0])
        for i in range(alive[0], alive[1]):
            if i<1500:
                temp[i] += 1
    immigrants.append(temp)

#  for j in range(1500):
#    for k in range(len(d)):
#        temp[j] +=(j in range(d['Birth'][k],d['Death'][k]+1))
#  timestamp.append(temp)
```

```python
print(len(immigrants[0]))
x = [[i in range(len(immigrants[j]))] for j in range(n)] 
for i in range(10):
    plt.plot(immigrants[i][:1500], color = "lightskyblue")
plt.plot([sum(values) / len(values) for values in zip(*immigrants)][:1500], color ="navy", linewidth = 2, linestyle='--', label = "Average")
plt.ylim(ymin=0)
plt.axvline(x = 495, color = 'orange', label = 'Solution Implemented')
plt.legend(loc="lower left")
plt.title('Immigration: Agent count over time to 1500 ticks')
```

```python
ts1 = [sum(values) / len(values) for values in zip(*basecase)][550:650]
ts2 =[sum(values) / len(values) for values in zip(*immigrants)][550:650]

# Determine midpoint
midpoint_index = len(ts1) // 2

# Calculate adjustment factor
adjustment_factor = (ts2[midpoint_index] - ts1[midpoint_index]) / 2

# Adjust values of time series 1 and 2
ts1 += adjustment_factor
ts2 -= adjustment_factor

# Plot the adjusted time series
import matplotlib.pyplot as plt

plt.plot(ts1, label='Time Series 1')
plt.plot(ts2, label='Time Series 2')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Adjusted Time Series')
plt.legend()
plt.show()

# Perform the Mann-Whitney U test
u_statistic, p_value = mannwhitneyu(ts1, ts2)
# Print the results
print("Mann-Whitney U test for the hump:")
print("U statistic:", u_statistic)
print("p-value:", p_value)
```

```python
from scipy.stats import mannwhitneyu

a0 =[sum(values) / len(values) for values in zip(*basecase)]
a1 =[sum(values) / len(values) - 200 for values in zip(*immigrants)]

plt.plot(a1[525:550], color ='steelblue', label ='Immigrants added')
plt.plot(a0[525:550], color = 'mediumpurple', label = 'Base case')

plt.ylim(ymin=0)
plt.legend(loc="lower left")
plt.title('Average of both 1500 ticks')
decay_hump = a0[525:550]  # Assuming the hump occurs in the first three points
hump_hump = a1[525:550]

# Perform the Mann-Whitney U test
u_statistic, p_value = mannwhitneyu(decay_hump, hump_hump)

# Print the results
print("Mann-Whitney U test for the hump:")
print("U statistic:", u_statistic)
print("p-value:", p_value)

# Check for significance at a chosen alpha level (e.g., alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("The case where pollution is removed is significantly different in the short term.")
else:
    print("The case where pollution is removed is not significantly different in the short term.")
```

Pollution

```python
n=9
data = []
timestamp = []
for i in range(11,n+11):
    data.append(pd.read_csv("https://raw.githubusercontent.com/javmencia/COBWEBfiles/main/{}.csv".format(i), sep="\t"))
```

```python
pollution = []
for i in range(n):
    print(i)
    d =data[i]
    temp = np.zeros(1500)

    for a in range(len(d)):
        if d['Death'][a]==-1:
            alive = [d['Birth'][a], 1500]
        elif d['Birth'][a]==-1:
            alive = [0, d['Death'][a]+1]
        else:
            alive = [d['Birth'][a], d['Death'][a]+1]
        incr = np.ones(max(0,alive[1]-alive[0]))
        for i in range(alive[0], alive[1]):
            if i<1500:
                temp[i] += 1
    pollution.append(temp)
```

```python
print(len(pollution[0]))
x = [[i in range(len(pollution[j]))] for j in range(n)] 
for i in range(n):
    plt.plot(pollution[i][:1500], color = "lightskyblue")
plt.plot([sum(values) / len(values) for values in zip(*pollution)][:1500], color ="navy", linewidth = 2, linestyle='--', label = "Average")
plt.ylim(ymin=0)
plt.axvline(x = 495, color = 'orange', label = 'Solution Implemented')

plt.legend(loc="lower left")
plt.title('Pollution: Agent count over time to 1500 ticks')
```

```python
ts1 = [sum(values) / len(values) for values in zip(*basecase)][500:600]
ts2 =[sum(values) / len(values) for values in zip(*pollution)][500:600]

# Determine midpoint
midpoint_index = len(ts1) // 2

# Calculate adjustment factor
adjustment_factor = (ts2[midpoint_index] - ts1[midpoint_index]) / 2

# Adjust values of time series 1 and 2
ts1 += adjustment_factor
ts2 -= adjustment_factor

# Plot the adjusted time series
import matplotlib.pyplot as plt

plt.plot(ts1, label='Time Series 1')
plt.plot(ts2, label='Time Series 2')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Adjusted Time Series')
plt.legend()
plt.show()

# Perform the Mann-Whitney U test
u_statistic, p_value = mannwhitneyu(ts1, ts2)
# Print the results
print("Mann-Whitney U test for the hump:")
print("U statistic:", u_statistic)
print("p-value:", p_value)
```

```python
from scipy.stats import pearsonr
ts1 = [sum(values) / len(values) for values in zip(*basecase)][500:1500]
ts2 =[sum(values) / len(values) for values in zip(*pollution)][500:1500]

# Calculate the Pearson correlation coefficient between the two time series
correlation_coefficient, p_value = pearsonr(ts1, ts2)

plt.plot(ts1, label='Time Series 1')
plt.plot(ts2, label='Time Series 2')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Adjusted Time Series')
plt.legend()
plt.show()
# Print the Pearson correlation coefficient
print("Pearson correlation coefficient:", correlation_coefficient)

# Interpret the coefficient: closer to 1 indicates more similar patterns
# You can set a threshold to determine if the patterns are significantly different
# For example:
threshold = 0.5  # Adjust the threshold based on your data and requirements
if correlation_coefficient > threshold:
    print("The overall patterns of the two time series are similar.")
else:
    print("The overall patterns of the two time series are significantly different.")
```
